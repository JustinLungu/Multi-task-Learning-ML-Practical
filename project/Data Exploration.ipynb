```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from matplotlib.ticker import MaxNLocator
```

```python
df = pd.read_csv('../data/UTKFace_labels.csv')
```

```python
df.head()
```

```python
df['Gender'].value_counts()
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed
df['Gender'].value_counts().plot(kind='bar', color='hotpink')

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Occurences of Gender in dataset')

labels = ['Male', 'Female']

# Show plot
plt.xticks(range(len(labels)), labels, rotation = 0)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines on y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed
plt.hist(df['Age'], bins=10, color='hotpink')  # Adjust bins as needed

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age Frequency')

# Show plot
plt.grid(True)
plt.show()
```

```python
plt.close()
```

```python
# Creating the boxplot
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed
plt.boxplot([df[df['Gender'] == 0]['Age'], df[df['Gender'] == 1]['Age']], labels=['Male', 'Female'])

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age by Gender')

# Show plot
plt.grid(True)
plt.show()
```

```python
plt.close()
```

```python
# Creating the boxplot
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed
plt.boxplot([df[df['Race'] == 0]['Age'], df[df['Race'] == 1]['Age'], df[df['Race'] == 2]['Age'], df[df['Race'] == 3]['Age'], df[df['Race'] == 4]['Age']], labels=['White', 'Black', 'Asian', 'Indian', 'Others'])

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age by Race')

# Show plot
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed
df['Race'].value_counts().plot(kind='bar', color='hotpink')

# Adding labels and title
plt.xlabel('Race')
plt.ylabel('Frequency')
plt.title('Occurences of Races in dataset')

labels = ['White', 'Black', 'Asian', 'Indian', 'Others']

# Show plot
plt.xticks(range(len(labels)), labels, rotation = 0)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines on y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
new_df = df.drop('Image_ID', axis = 'columns')
```

```python
tsne_2d = TSNE(n_components=2)
df_tsne = tsne_2d.fit_transform(new_df)
```

```python
# Plot t-SNE results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
scatter = plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c = df['Age'], cmap=plt.cm.get_cmap('viridis', 10))
plt.title('t-SNE (2D)')

cbar = plt.colorbar(scatter)
cbar.set_label('Age')

plt.show()
```

```python
# Plot t-SNE results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
scatter = plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c = df['Race'], cmap=plt.cm.get_cmap('viridis', 10))
plt.title('t-SNE (2D)')

cbar = plt.colorbar(scatter)
cbar.set_label('Race')
cbar.locator = MaxNLocator(nbins=4)

plt.show()
```

```python
# Plot t-SNE results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
scatter = plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c = df['Gender'], cmap=plt.cm.get_cmap('viridis', 10))
plt.title('t-SNE (2D)')

cbar = plt.colorbar(scatter)
cbar.set_label('Gender')
cbar.locator = MaxNLocator(nbins=1)

plt.show()
```
